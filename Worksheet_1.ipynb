# **Problem - 1: Array Creation:**
Complete the following Tasks:
1. Initialize an empty array with size 2X2.
2. Initialize an all one array with size 4X2.
3. Return a new array of given shape and type, filled with fill value.{Hint: np.full}
4. Return a new array of zeros with same shape and type as a given array.{Hint: np.zeros like}
5. Return a new array of ones with same shape and type as a given array.{Hint: np.ones like}
6. For an existing list new_list = [1,2,3,4] convert to an numpy array.{Hint: np.array()}

```python
import numpy as np

# 1. Initialize an empty array with size 2X2.
empty_array = np.empty((2, 2))
print("1. Empty Array (2x2):\n", empty_array)

# 2. Initialize an all one array with size 4X2.
ones_array = np.ones((4, 2))
print("\n2. All Ones Array (4x2):\n", ones_array)

# 3. Return a new array of given shape and type, filled with fill value.
fill_value_array = np.full((3, 3), 7)  # 3x3 array filled with 7
print("\n3. Filled Array (3x3 with 7):\n", fill_value_array)

# 4. Return a new array of zeros with same shape and type as a given array.
sample_array = np.array([[1, 2], [3, 4]])
zeros_like_array = np.zeros_like(sample_array)
print("\n4. Zeros Like Array:\n", zeros_like_array)

# 5. Return a new array of ones with same shape and type as a given array.
ones_like_array = np.ones_like(sample_array)
print("\n5. Ones Like Array:\n", ones_like_array)

# 6. Convert list to numpy array.
new_list = [1, 2, 3, 4]
numpy_array = np.array(new_list)
print("\n6. Converted Numpy Array:\n", numpy_array)
```

# **Problem - 2: Array Manipulation: Numerical Ranges and Array indexing:**
Complete the following tasks:
1. Create an array with values ranging from 10 to 49. {Hint:np.arrange()}.
2. Create a 3X3 matrix with values ranging from 0 to 8.
{Hint:look for np.reshape()}
3. Create a 3X3 identity matrix.{Hint:np.eye()}
4. Create a random array of size 30 and find the mean of the array.
{Hint:check for np.random.random() and array.mean() function}
5. Create a 10X10 array with random values and find the minimum and maximum values.
6. Create a zero array of size 10 and replace 5th element with 1.
7. Reverse an array arr = [1,2,0,0,4,0].
8. Create a 2d array with 1 on border and 0 inside.
9. Create a 8X8 matrix and fill it with a checkerboard pattern.

```python
# 1. Create an array with values ranging from 10 to 49.
range_array = np.arange(10, 50)
print("\n1. Array with values 10 to 49:\n", range_array)

# 2. Create a 3x3 matrix with values ranging from 0 to 8.
reshaped_array = np.arange(9).reshape(3, 3)
print("\n2. 3x3 Matrix:\n", reshaped_array)

# 3. Create a 3x3 identity matrix.
identity_matrix = np.eye(3)
print("\n3. Identity Matrix (3x3):\n", identity_matrix)

# 4. Create a random array of size 30 and find the mean.
random_array = np.random.random(30)
mean_value = random_array.mean()
print("\n4. Random Array Mean:\n", mean_value)

# 5. Create a 10x10 array with random values and find min and max values.
large_random_array = np.random.random((10, 10))
min_value = large_random_array.min()
max_value = large_random_array.max()
print("\n5. Min and Max of 10x10 Random Array:\nMin:", min_value, "Max:", max_value)

# 6. Create a zero array of size 10 and replace 5th element with 1.
zero_array = np.zeros(10)
zero_array[4] = 1
print("\n6. Zero Array with 5th Element as 1:\n", zero_array)

# 7. Reverse an array arr = [1, 2, 0, 0, 4, 0].
arr = [1, 2, 0, 0, 4, 0]
reversed_array = arr[::-1]
print("\n7. Reversed Array:\n", reversed_array)

# 8. Create a 2D array with 1 on border and 0 inside.
border_array = np.ones((5, 5))
border_array[1:-1, 1:-1] = 0
print("\n8. 2D Array with 1 Border and 0 Inside:\n", border_array)

# 9. Create an 8x8 matrix and fill it with a checkerboard pattern.
checkerboard = np.zeros((8, 8), dtype=int)
checkerboard[1::2, ::2] = 1  # 1 on odd rows, even columns
checkerboard[::2, 1::2] = 1  # 1 on even rows, odd columns
print("\n9. Checkerboard Pattern (8x8):\n", checkerboard)
```

{Hint: First use Numpy array to represent the equation in Matrix form. Then Solve for: AX = B}
‚Ä¢ Now: solve the above equation using np.linalg.inv function.{Explore more about ‚Äùlinalg‚Äù function
of Numpy}

```python
#Given
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

#1. Prove ùê¥‚ãÖùê¥‚àí1=ùêºA‚ãÖA ‚àí1 = I
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)
print("1. Proving A¬∑A‚Åª¬π = I:\n", identity_matrix)

#2. Prove ùê¥ùêµ‚â†ùêµùê¥
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("Are AB and BA equal? ", np.array_equal(AB, BA))

#3. Prove (ùê¥ùêµ)ùëá=ùêµùëá‚ãÖùê¥ùëá(AB)T =BT ‚ãÖ AT
AB_transpose = np.transpose(np.dot(A, B))
BT_AT = np.dot(B.T, A.T)
print("\n3. Proving (AB)^T = B^T¬∑A^T:\n", AB_transpose)
print("\nB^T¬∑A^T:\n", BT_AT)

#2x‚àí3y+z=‚àí1
#x‚àíy+2z=‚àí3
#3x+y‚àíz=9
#Represent in matrix form
# AX=B:
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

#Solve using np.linalg.inv:
A_inv = np.linalg.inv(A)
X = np.dot(A_inv, B)
print("\nSolution using inverse method:\n", X)

#Solve using NumPy:
X_direct = np.linalg.solve(A, B)
print("\nSolution using np.linalg.solve:\n", X_direct)

```
